```python
import matplotlib
import pandas as pd
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os


myfont = fm.FontProperties(fname=r'C:\Windows\Fonts\Times New Roman.tiff') # 设置字体

# 设置matplotlib正常显示中文和负号


# matplotlib.rcParams['font.sans-serif']=['NSimSun']   # 用黑体显示中文
# matplotlib.rcParams['axes.unicode_minus']=False     # 正常显示负号
# matplotlib.rcParams['font.size'] = 0.3
# matplotlib.rcParams['font.family'] = 'sans-serif'  
matplotlib.rcParams['font.sans-serif'] = 'Times New Roman'  # NSimSun 新宋体字体   宋体：SimSuncss

os.chdir(r'E:\BaiduSyncdisk\热浪博士论文\计算表格\Chapter4')   

colors = [ '#E1E1E1', '#66CDAB', '#F86F64']   


alldata = pd.read_csv('Reclassify_Z_exposure.csv')

print(alldata)
```

```python
data = alldata["historical"]

# plt.rcParams["font.weight"] = "bold"
fig, axe = plt.subplots(figsize=(1, 1), dpi=300)
# 调整图片的边距
# plt.subplots_adjust(left=0.025, right=0.945, top=0.99, bottom=0.0009)
wedges, texts, autotexts= plt.pie(data,colors = colors,pctdistance = 1.2,autopct='%.0f%%')
plt.setp(texts, size=10, weight='bold', color='black')

plt.savefig('00 TM_historical_2.png', dpi=300,facecolor="none", transparent=True)
plt.show()
```
