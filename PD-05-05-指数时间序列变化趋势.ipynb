```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 加载 Excel 文件
file_path = 'hazard_exposure_adaptation_index_results-hazard1-AGE.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')


# 去掉列名中的多余空格
df.columns = df.columns.str.strip()

# 定义模型中的系数
alpha = 0.10  # Hazard Index 系数
beta = 0.36   # Exposure Index 系数
gamma = 0.54  # Adaptation Index 系数

# 提取计算风险指数所需的列
df_selected = df[['年份', '地区', 'Hazard Index', 'Exposure Index', 'Adaptation Index']]

# 选择对数变换，控制风险指数的变化
df_selected['Risk Index'] = 1 / (1 + np.exp(
    - np.log(alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index']) + 1)
))

# 按年份分组，计算每个省份的均值
df_mean_by_province = df_selected.groupby('年份')[['Adaptation Index', 'Exposure Index', 'Hazard Index', 'Risk Index']].mean().reset_index()

# 显示数据框，包含均值
print(df_mean_by_province.head())

# 定义线性回归函数
def lm_func(df, col):
    """简单的时间序列线性回归，返回系数、R2和p值"""
    X_year = np.array(df["年份"]).reshape((-1, 1))
    Y_stats = np.array(df[col]).reshape((-1, 1))

    # 加入截距项
    X_year_2 = sm.add_constant(X_year)

    # 执行回归
    model = sm.OLS(Y_stats, X_year_2).fit()

    coef = model.params[1]
    r2 = model.rsquared_adj
    p = model.pvalues[1]

    return round(coef, 4), round(r2, 2), round(p, 3)

# 为每个指数应用线性回归
coef_adaptation, r2_adaptation, p_adaptation = lm_func(df_mean_by_province, 'Adaptation Index')
coef_exposure, r2_exposure, p_exposure = lm_func(df_mean_by_province, 'Exposure Index')
coef_hazard, r2_hazard, p_hazard = lm_func(df_mean_by_province, 'Hazard Index')
coef_health_risk, r2_health_risk, p_health_risk = lm_func(df_mean_by_province, 'Risk Index')

# 添加回归线（使用sns.regplot）
sns.regplot(data=df_mean_by_province, x='年份', y='Adaptation Index', scatter=False, color='g', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Adaptation Index', label='Adaptation Index', marker='o', color='g')

sns.regplot(data=df_mean_by_province, x='年份', y='Exposure Index', scatter=False, color='purple', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Exposure Index', label='Exposure Index', marker='o', color='purple')

sns.regplot(data=df_mean_by_province, x='年份', y='Hazard Index', scatter=False, color='orange', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Hazard Index', label='Hazard Index', color='orange', marker='o')

sns.regplot(data=df_mean_by_province, x='年份', y='Risk Index', scatter=False, color='r', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Risk Index', label='Health Risk Index', color='r', marker='o')

# 添加标题和标签
plt.ylabel('Index Value', fontsize=22)
plt.xlabel(' ', fontsize=22)

# 显示图例
plt.legend(fontsize=14, bbox_to_anchor=(0.1, 1.01, 0.9, 1.12), loc='lower left',
           handletextpad=0.6, ncol=4, mode="expand", borderaxespad=0., handlelength=1.5)

# 显示图像
plt.tight_layout()

# 保存图像
plt.savefig(r'E:\BaiduSyncdisk\热浪博士论文\论文图片\PD-05-04-风险指数变化-0901.png', dpi=400, bbox_inches='tight', facecolor="none", transparent=True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt

# 加载 Excel 文件
file_path = 'hazard_exposure_adaptation_index_results-hazard1-AGE.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# 去掉列名中的多余空格
df.columns = df.columns.str.strip()

# 定义模型中的系数
alpha = 0.15  # 增大Hazard Index 系数
beta = 0.50   # 增大Exposure Index 系数
gamma = 0.30  # 增大Adaptation Index 系数

# 提取计算风险指数所需的列
df_selected = df[['年份', '地区', 'Hazard Index', 'Exposure Index', 'Adaptation Index']]

# 选择对数变换，增强风险指数的变化
# 使用指数变换，使波动更强烈
df_selected['Risk Index'] = 1 / (1 + np.exp(
    - (alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index'])) ** 2
))

# 按年份分组，计算每个省份的均值
df_mean_by_province = df_selected.groupby('年份')[['Adaptation Index', 'Exposure Index', 'Hazard Index', 'Risk Index']].mean().reset_index()

# 显示数据框，包含均值
print(df_mean_by_province.head())

# 定义线性回归函数
def lm_func(df, col):
    """简单的时间序列线性回归，返回系数、R2和p值"""
    X_year = np.array(df["年份"]).reshape((-1, 1))
    Y_stats = np.array(df[col]).reshape((-1, 1))

    # 加入截距项
    X_year_2 = sm.add_constant(X_year)

    # 执行回归
    model = sm.OLS(Y_stats, X_year_2).fit()

    coef = model.params[1]
    r2 = model.rsquared_adj
    p = model.pvalues[1]

    return round(coef, 4), round(r2, 2), round(p, 3)

# 为每个指数应用线性回归
coef_adaptation, r2_adaptation, p_adaptation = lm_func(df_mean_by_province, 'Adaptation Index')
coef_exposure, r2_exposure, p_exposure = lm_func(df_mean_by_province, 'Exposure Index')
coef_hazard, r2_hazard, p_hazard = lm_func(df_mean_by_province, 'Hazard Index')
coef_health_risk, r2_health_risk, p_health_risk = lm_func(df_mean_by_province, 'Risk Index')

# 添加回归线（使用sns.regplot）
sns.regplot(data=df_mean_by_province, x='年份', y='Adaptation Index', scatter=False, color='g', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Adaptation Index', label='Adaptation Index', marker='o', color='g')

sns.regplot(data=df_mean_by_province, x='年份', y='Exposure Index', scatter=False, color='purple', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Exposure Index', label='Exposure Index', marker='o', color='purple')

sns.regplot(data=df_mean_by_province, x='年份', y='Hazard Index', scatter=False, color='orange', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Hazard Index', label='Hazard Index', color='orange', marker='o')

sns.regplot(data=df_mean_by_province, x='年份', y='Risk Index', scatter=False, color='r', line_kws={'linewidth': 1})
sns.lineplot(data=df_mean_by_province, x='年份', y='Risk Index', label='Health Risk Index', color='r', marker='o')

# 添加标题和标签
plt.ylabel('Index Value', fontsize=22)
plt.xlabel(' ', fontsize=22)

# 显示图例
plt.legend(fontsize=14, bbox_to_anchor=(0.1, 1.01, 0.9, 1.12), loc='lower left',
           handletextpad=0.6, ncol=4, mode="expand", borderaxespad=0., handlelength=1.5)

# 显示图像
plt.tight_layout()

# 保存图像
plt.savefig(r'E:\BaiduSyncdisk\热浪博士论文\论文图片\PD-05-04-风险指数变化-0901.png', dpi=400, bbox_inches='tight', facecolor="none", transparent=True)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 加载 Excel 文件
file_path = 'hazard_exposure_adaptation_index_results-hazard1-AGE.xlsx'
df = pd.read_excel(file_path, sheet_name='Sheet1')

# 去掉列名中的多余空格
df.columns = df.columns.str.strip()

# 定义模型中的系数
alpha = 0.10  # Hazard Index 系数
beta = 0.36   # Exposure Index 系数
gamma = 0.54  # Adaptation Index 系数

# 提取计算风险指数所需的列
df_selected = df[['年份', '地区', 'Hazard Index', 'Exposure Index', 'Adaptation Index']]

# 方案 1: 使用平方变换
df_selected['Risk Index_Squared'] = 1 / (1 + np.exp(
    - (alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index'])) ** 2
))

# 方案 2: 使用指数变换
df_selected['Risk Index_Exponential'] = 1 / (1 + np.exp(
    - (alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index'])) * np.exp(alpha * df_selected['Hazard Index'])
))

# 方案 3: 使用对数变换
df_selected['Risk Index_Log'] = 1 / (1 + np.exp(
    - np.log(alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index']) + 1)
))

# 方案 4: 使用对数和平方组合变换
df_selected['Risk Index_Log_Squared'] = 1 / (1 + np.exp(
    - np.log(alpha * df_selected['Hazard Index'] + beta * df_selected['Exposure Index'] + gamma * (1 - df_selected['Adaptation Index']) + 1)**2
))

# 按年份分组，计算每个省份的均值
df_mean_by_province = df_selected.groupby('年份')[['Risk Index_Squared', 'Risk Index_Exponential', 'Risk Index_Log', 'Risk Index_Log_Squared']].mean().reset_index()

# 显示计算后的结果
print(df_mean_by_province.head())

# 可选：将结果保存为新的 Excel 文件
df_mean_by_province.to_excel('calculated_risk_index_comparison.xlsx', index=False)

# 绘制风险指数趋势图
plt.figure(figsize=(10, 6))
plt.plot(df_mean_by_province['年份'], df_mean_by_province['Risk Index_Squared'], label='Squared Transformation', color='b')
plt.plot(df_mean_by_province['年份'], df_mean_by_province['Risk Index_Exponential'], label='Exponential Transformation', color='g')
plt.plot(df_mean_by_province['年份'], df_mean_by_province['Risk Index_Log'], label='Logarithmic Transformation', color='r')
plt.plot(df_mean_by_province['年份'], df_mean_by_province['Risk Index_Log_Squared'], label='Log and Squared Transformation', color='orange')

plt.xlabel('Year', fontsize=14)
plt.ylabel('Risk Index', fontsize=14)
plt.title('Risk Index Trends with Different Transformations', fontsize=16)
plt.legend()
plt.tight_layout()
plt.show()
```
